### The next step

```python
random_walk = [0]
for i in range(1,101):
    step = random_walk[-1]
    dice = np.random.randint(1,7)
    if dice == 1 or dice == 2:
        step -= 1
    elif dice == 3 or dice == 4 or dice == 5:
        step += 1
    else:
        dice = np.random.randint(1,7)
        step += dice
    random_walk.append(step)
print(random_walk)
```

### How long can you go?

```python
np.random.seed(123)
random_walk = [0]
for i in range(1,101):
    step = random_walk[-1]
    dice = np.random.randint(1,7)
    if dice <= 2:
        step = max(0, step-1)
    elif dice > 2 or dice < 6:
        step += 1
    else:
        dice= np.random.randint(1,7)
        step += dice
    random_walk.append(step)
print(random_walk)
```

### Visualize the walk

```python
plt.plot(random_walk)
plt.show()
```

### Simulate multiple walks

```python
all_walks=[]
for k in range(1,11):
    random_walk = [0]
    for i in range(1,101):
        step = random_walk[-1]
        dice = np.random.randint(1,7)
        if dice <= 2:
            step = max(0, step-1)
        elif dice > 2 or dice < 6:
            step += 1
        else:
            dice= np.random.randint(1,7)
            step += dice
        random_walk.append(step)
    all_walks.append(random_walk)
print(all_walks)
```

### Visualize all walks

```python
np_aw = np.array(all_walks)
plt.plot(np_aw)
plt.show()
```

```python
np_aw_t = np.transpose(np_aw)
np_aw_t
```

```python
plt.figure(figsize=(12,6))
plt.plot(np_aw_t)
plt.show()
```
