# Problem 4: Sector Analysis 1

```python
mapping=pd.read_csv('C:/Users/Subhadri/Desktop/Data Science/Term 1/Pandas/Assignment 1 - Teclov/mapping.csv')
mapping.head()
```

```python
# 1. Extract the primary sector of each category list from the category_list column.
list1=[]
dic={'category_list': str,}
# dic
master_frame['category_list']=master_frame['category_list'].astype(dic)
cat_list=master_frame['category_list']
#cat_list
for i in cat_list:
    i=i.split('|')
    list1.append(i[0])
# list1
primary_sector=pd.DataFrame(list1,columns=['category_list'])
primary_sector.head()
```

```python
# 2. Use the mapping file 'mapping.csv' to map each primary sector to one of the eight main sectors (Note that ‘Others’ is also 
# considered one of the main sectors)
k=pd.melt(mapping,id_vars='category_list',var_name='Main_Sector')
k=k[['category_list','Main_Sector']][k.value==1].sort_values('category_list')
#k.head()
sector=pd.merge(primary_sector,k,on='category_list')
#sector.shape
sector=sector.drop_duplicates()
#sector.shape
sector.rename(columns={'category_list': 'Primary_Sector'},inplace=True)
sector.head()
```

# Problem 5: Sector Analysis 2

```python
master_sector=pd.merge(sector,master_frame,left_on='Primary_Sector',right_on='category_list')
master_sector.tail()
```

```python
master_sector=master_sector.rename(columns={'funding_round_type':'FT'})
master_sector.head()
```

```python
Country1=master_sector[(master_sector.country_code=='USA') & (master_sector.FT=='venture') & 
                       ((master_sector.raised_amount_usd >=5000000) | (master_sector.raised_amount_usd<=15000000))]
#print(Country1)
Country2=master_sector[(master_sector.country_code=='IND') & (master_sector.FT=='venture') & 
                       ((master_sector.raised_amount_usd >=5000000) | (master_sector.raised_amount_usd<=15000000))]
#print(Country2)
Country3=master_sector[(master_sector.country_code=='GBR') & (master_sector.FT=='venture') & 
                       ((master_sector.raised_amount_usd >=5000000) | (master_sector.raised_amount_usd<=15000000))]
#print(Country3)
```

```python
total_amount1=Country1.groupby('Main_Sector')['raised_amount_usd'].sum()
#print(total_amount1)
p1=Country1.groupby('Main_Sector').size()
#print(p1)
df1=pd.merge(Country1,total_amount1,on='Main_Sector', suffixes=('','Total'))
df1.head()
total_no_investment1=pd.DataFrame(p1,columns=['total_counts'])
total_no_investment1
```

```python
D1=pd.merge(df1,total_no_investment1,on='Main_Sector')
D1.head()
```
